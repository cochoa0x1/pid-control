```python
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_csv('build/sample7.csv',  sep=' ',header=None, skiprows=3)
```

```python
%pylab inline
```

```python
df=df.dropna()
```

```python
col_names = [df[c][0] for i,c in enumerate(df.columns) if i%2==0]
col_names = map(lambda x: x.replace(':',''),col_names)
```

```python
dfx = df[[c for i,c in enumerate(df.columns) if i%2==1]]
dfx.columns = list(col_names)
```

```python
figure(figsize(10,5))
dfx.cte.plot()
ylim(-5,5)


figure(figsize(10,5))
dfx.cte.diff().plot()
ylim(-1,1)

figure(figsize(10,5))
dfx.cte.cumsum().plot()

figure(figsize(10,5))
dfx[['de','pe','ie']].plot()
ylim(-1,1)
```

```python
dfx[['de','pe','ie']].plot(kind='kde')
xlim(-1,1)
```
